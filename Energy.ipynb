```python
import matplotlib.pyplot as plt

# Example energy consumption data (in MWh)
energy_sources_mwh = {
    'Purchased Electricity': 5000,
    'Diesel': 1500,
    'Gasoline': 800,
    'Natural Gas': 300,
    'Solar': 200,
    'Other Renewables': 100
}

# Convert MWh to MJ (1 MWh = 3600 MJ)
energy_sources_mj = {k: v * 3600 for k, v in energy_sources_mwh.items()}

# Colors for the doughnut
colors = ['#66b3ff','#ff9999','#99ff99','#ffcc99','#c2c2f0','#ffb3e6']

# Plotting the doughnut chart
fig, ax = plt.subplots(figsize=(8, 8))

values = list(energy_sources_mj.values())
labels = list(energy_sources_mj.keys())

# Create doughnut
wedges, texts = ax.pie(values, startangle=90, colors=colors, wedgeprops=dict(width=0.4, edgecolor='white'))

# Add labels outside with leader lines
ax.legend(wedges, [f"{label}: {value:,} MJ" for label, value in energy_sources_mj.items()],
          title="Energy Sources", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Total energy in the center
total_energy = sum(values)
ax.text(0, 0, f'Total\n{total_energy:,} MJ', ha='center', va='center', fontsize=14, fontweight='bold')

# Title
ax.set_title('XYZ Energy Consumption Report (GRI 302)', fontsize=16)

plt.tight_layout()
plt.show()
```
