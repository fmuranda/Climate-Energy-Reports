```python
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt

# 1. Prepare Data
# Define the 15 categories for Scope 3
scope3_categories = [
    'Purchased goods and services',
    'Capital goods',
    'Fuel- and energy-related activities',
    'Upstream transportation and distribution',
    'Waste generated in operations',
    'Business travel',
    'Employee commuting',
    'Upstream leased assets',
    'Downstream transportation and distribution',
    'Processing of sold products',
    'Use of sold products',
    'End-of-life treatment of sold products',
    'Downstream leased assets',
    'Franchises',
    'Investments'
]

# Distribute the original Scope 3 emissions (120000.0) evenly across 15 categories for now
emissions_per_scope3_category = 120000.0 / 15

data = {
    'Category': [
        'Scope 1 (Direct)',
        'Scope 2 (Indirect)'
    ] + [f'Scope 3 - {cat}' for cat in scope3_categories],
    'Emissions': [
        54649.3,
        39091.4
    ] + [emissions_per_scope3_category] * 15,
    'Description': [
        'Diesel & Fleet',
        'Electricity usage'
    ] + [f'Details for {cat}' for cat in scope3_categories]
}
df = pd.DataFrame(data)

# Define Econet-style Branding colors for the donut chart
econet_colors = ["#0056b3", "#28a745"] + px.colors.sequential.Sunset[:15] # Using more colors for Scope 3

# 2. Create a Donut Chart (Modern for Annual Reports)
fig = px.pie(df, values='Emissions', names='Category', hole=0.5,
             title='Econet Wireless GHG Emissions Profile (2025) - Detailed Scope 3',
             color_discrete_sequence=econet_colors,
             hover_data=['Description'])

# 3. Update Layout for "Executive" look
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(
    font_family="Arial",
    title_font_size=24,
    showlegend=False,
    margin=dict(t=80, b=40, l=40, r=40)
)

fig.show()
plt.savefig('Econet Wireless GHG Emissions_profile.png', dpi=300)
```

```python
import seaborn as sns


# Setup aesthetics
sns.set_theme(style="whitegrid")
econet_colors = ["#0056b3", "#28a745", "#ffc107"] # Original Econet-style Branding (not used directly for this plot)

# Data for Year-on-Year Trend
trend_data = pd.DataFrame({
    'Year': ['2024', '2024', '2025', '2025'],
    'Scope': ['Scope 1', 'Scope 2', 'Scope 1', 'Scope 2'],
    'Emissions': [45000, 32000, 54649.3, 39091.4]
})

# Create Plot
plt.figure(figsize=(8, 6))

# Define the new palette: blue for Scope 1, gray for Scope 2
bar_line_palette = ["#0056b3", "#808080"]

plot = sns.barplot(data=trend_data, x='Year', y='Emissions', hue='Scope', palette=bar_line_palette)

# Add the dotted trend line on top of the bar chart
sns.lineplot(data=trend_data, x='Year', y='Emissions', hue='Scope',
             palette=bar_line_palette, linestyle=':', marker='o', ax=plot.axes, legend=False)

# Customizing for a "Report" look
plt.title('Greenhouse Gas Emissions Trend (Scope 1 & 2)', fontsize=16, fontweight='bold', pad=20)
plt.ylabel('Emissions (tCO$_2$e)', fontsize=12)
plt.xlabel('Financial Year', fontsize=12)
sns.despine(left=True)

# Add value labels on top of bars with units
for p in plot.patches:
    plot.annotate(f"{p.get_height():.1f} tCO$_2$e",
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha = 'center', va = 'center',
                   xytext = (0, 9),
                   textcoords = 'offset points')

plt.tight_layout()
plt.savefig('econet_emissions_report.png', dpi=300)
```

# Task
Extract and visualize the distribution of emissions across the 15 individual Scope 3 categories from the existing DataFrame, and provide a summary of the distribution, highlighting key observations.

## Extract Scope 3 Data

### Subtask:
Filter the existing DataFrame to isolate only the Scope 3 emission categories and their corresponding emission values.


**Reasoning**:
Filter the existing DataFrame `df` to extract only the Scope 3 categories and their emission values, storing them in a new DataFrame called `df_scope3`.



```python
df_scope3 = df[df['Category'].str.startswith('Scope 3 -')][['Category', 'Emissions']]
print(df_scope3.head())
```

**Reasoning**:
Now that the Scope 3 data has been extracted into `df_scope3`, the next step is to visualize the distribution of these emissions across the 15 categories. A bar chart is an effective way to show this comparison.



```python
import plotly.express as px

fig = px.bar(df_scope3,
             x='Category',
             y='Emissions',
             title='Distribution of GHG Emissions Across 15 Scope 3 Categories',
             labels={'Category': 'Scope 3 Category', 'Emissions': 'Emissions (tCO2e)'},
             color_discrete_sequence=px.colors.qualitative.Plotly)

fig.update_layout(
    xaxis_tickangle=-45, # Angle x-axis labels for readability
    font_family="Arial",
    title_font_size=20,
    showlegend=False,
    margin=dict(t=80, b=100, l=40, r=40) # Adjust margins for labels
)

fig.show()
plt.savefig('Distributiono of GHG Emissions Accross 15 Scope 3 Catgories.png', dpi=300)
```

### Summary of Scope 3 Emissions Distribution

The visualization of the 15 Scope 3 categories reveals that, in this dataset, the emissions are **evenly distributed** across all categories. Each of the 15 Scope 3 categories accounts for exactly 8000 tCO2e of emissions.

**Key Observations:**
*   **Equal Contribution:** Every Scope 3 category contributes an identical amount (8000 tCO2e) to the total Scope 3 emissions.
*   **Initial Distribution:** This uniform distribution indicates that, for the purpose of this initial analysis, a simplified approach was taken where the total Scope 3 emissions were divided equally among its constituent categories. This is a common starting point in GHG accounting when detailed data for each sub-category is not yet available, or for illustrative purposes.
