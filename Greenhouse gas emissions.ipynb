```python
import matplotlib.pyplot as plt

# Example data for XYZ GHG emissions (in metric tons CO2e)
emissions = {
    'Scope 1': {'Stationary Combustion': 1200, 'Mobile Combustion': 800, 'Fugitive Emissions': 200},
    'Scope 2': {'Purchased Electricity': 5000, 'Purchased Steam': 300},
    'Scope 3': {'Business Travel': 600, 'Waste': 150, 'Purchased Goods & Services': 3500, 'Upstream Transport': 400}
}

# Colors for breakdowns
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6','#c2f0c2','#ff6666']

# Plotting doughnut charts
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for ax, (scope, breakdown) in zip(axes, emissions.items()):
    values = list(breakdown.values())
    labels = list(breakdown.keys())
    
    wedges, texts = ax.pie(values, labels=labels, startangle=90, colors=colors[:len(values)],
                           wedgeprops=dict(width=0.4))
    
    ax.set_title(f'{scope} GHG Emissions', fontsize=14)
    
# Central title
fig.suptitle('XYZ GHG Emissions Breakdown by Scope', fontsize=16)
plt.tight_layout()
plt.show()
```
